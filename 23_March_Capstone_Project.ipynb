```python

#  Build an End-to-End RAG + Agent System


from sentence_transformers import SentenceTransformer
import faiss
import numpy as np
from transformers import pipeline

# =========================
# 📄 DOCUMENTS
# =========================
documents = [
    "Refund policy allows refund within 7 days.",
    "Assignments must be submitted before Sunday.",
    "Lecture notes are available as PDFs and videos.",
    "Course includes AI, ML, and Deep Learning topics.",
    "RAG improves factual accuracy by retrieving documents.",
    "Agents enable decision making and tool usage."
]

# =========================
# CHUNKING
# =========================
def chunk_text(text, size=50):
    words = text.split()
    return [" ".join(words[i:i+size]) for i in range(0, len(words), size)]

chunks = []
for doc in documents:
    chunks.extend(chunk_text(doc))

# =========================
#  EMBEDDINGS
# =========================
embedder = SentenceTransformer('all-MiniLM-L6-v2')
embeddings = embedder.encode(chunks)

# =========================
# VECTOR STORE
# =========================
dim = embeddings.shape[1]
index = faiss.IndexFlatL2(dim)
index.add(np.array(embeddings))

# =========================
# RETRIEVAL
# =========================
def retrieve(query, k=2):
    q_embed = embedder.encode([query])
    distances, indices = index.search(np.array(q_embed), k)
    return [chunks[i] for i in indices[0]]

# =========================
# TOOL
# =========================
def get_student_status(student_id):
    return f"Student {student_id} is enrolled and completed 70% course."

# =========================
# LLM (FIXED SETTINGS)
# =========================
generator = pipeline("text-generation", model="distilgpt2")

# =========================
# AGENT
# =========================
def agent(query):

    # TOOL decision
    if any(word in query.lower() for word in ["status", "progress", "enrollment"]):
        data = get_student_status("101")
        return f"📊 {data}"

    # RAG
    context = retrieve(query)

    # If no useful context → safe answer
    if not context:
        return "❗ I don't know based on available data."

    # Simple deterministic answer (NO GPT confusion)
    return f"📘 Based on documents:\n- " + "\n- ".join(context)


# =========================
# INTERACTIVE LOOP
# =========================
print("🤖 EdTech AI Assistant Ready! (type 'exit' to stop)\n")

while True:
    q = input("👤 Ask your question: ")

    if q.lower() == "exit":
        print("👋 Goodbye!")
        break

    print("\n🎯 ANSWER:\n", agent(q))
    print("\n" + "="*50)
```

```python

#  INTERACTIVE MULTI-AGENT SYSTEM (FINAL VERSION)


from transformers import pipeline

# =========================
# LLM
# =========================
generator = pipeline("text-generation", model="distilgpt2")

# =========================
# KNOWLEDGE BASE
# =========================
knowledge_base = {
    "transaction": "Transactions are processed instantly and securely.",
    "fraud": "Fraud transactions are detected using AI monitoring systems.",
    "refund": "Refunds are processed within 5-7 business days.",
    "deadline": "Assignments must be submitted before Sunday.",
    "faq": "You can contact support anytime for help."
}

# =========================
# RETRIEVAL AGENT
# =========================
def retrieval_agent(query):
    context = []

    for key in knowledge_base:
        if key in query.lower():
            context.append(knowledge_base[key])

    if not context:
        context.append("General support: please contact help desk.")

    return " ".join(context)


# =========================
# REASONING AGENT
# =========================
def reasoning_agent(query, context):

    prompt = f"""
    Use the context to answer clearly.

    Context: {context}
    Question: {query}
    """

    output = generator(
        prompt,
        max_new_tokens=80,
        pad_token_id=50256
    )[0]["generated_text"]

    return output


# =========================
# VALIDATION AGENT
# =========================
def validation_agent(answer):

    if len(answer) > 40:
        return True
    return False


# =========================
# MULTI-AGENT WORKFLOW
# =========================
def multi_agent_system(query):

    attempts = 0

    while attempts < 3:
        print(f"\n🔄 Attempt {attempts+1}")

        # Step 1: Retrieval
        context = retrieval_agent(query)
        print("📄 Context:", context)

        # Step 2: Reasoning
        answer = reasoning_agent(query, context)
        print("🧠 Answer Generated")

        # Step 3: Validation
        is_valid = validation_agent(answer)
        print("✅ Valid:", is_valid)

        if is_valid:
            return answer

        # Retry with improved query
        query += " explain clearly"
        attempts += 1
        print("❌ Retrying...\n")

    return "⚠️ Unable to generate proper answer."


# =========================
# INTERACTIVE CHAT LOOP
# =========================
print("🤖 Fintech AI Support System Ready!")
print("Type 'exit' to stop\n")

while True:

    user_input = input("👤 Ask your question: ")

    if user_input.lower() == "exit":
        print("👋 Goodbye!")
        break

    final_answer = multi_agent_system(user_input)

    print("\n🎯 FINAL ANSWER:\n", final_answer)
    print("\n" + "="*60)
```
